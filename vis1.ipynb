#  STOCKS COMPARISON

```python
i = int(input("how many stock to compare : "))
intrv=int(input("1. WEEKLY 2. INTRADAY"))
l=[]
for k in range(i):
    s1=input('enter symbol : ')
    if(intrv==1):
        fd=ts.get_weekly(symbol=s1)
        fd=fd[0].head(365)
        fd=fd[['4. close']]
        fd.columns=[s1]
        l.append(fd)
    elif(intrv==2):
        fd=ts.get_intraday(symbol=s1, interval='5min')
        fd=fd[0]
        fd=fd[['4. close']]
        fd.columns=[s1]
        l.append(fd)


ffd=pd.concat(l,axis=1)
    
    
    
```
